# this notebook analyzes NetRunner cards
## They are imported as a json file from netrunnerdb.com

```python
import pandas as pd
```

```python
cards = pd.read_json('allcards.json')
```

```python
# get an array of dictionaries from the 'data' column
cardarr = cards['data'].values
```

```python
factions = list({card['faction_code'] for card in cardarr})
print(factions)
[card['title'] for card in cardarr if card['faction_code']==factions[8] ]
```

# now let's examine the cards in the core sets

```python
coreset = {card['title'] for card in cardarr if card['pack_code']=='core'}
core2set = {card['title'] for card in cardarr if card['pack_code']=='core2'}
```

```python
# cards from the core set that didn't make it to core 2.0 (they rotated out)
coreset - core2set
```

```python
# cards that were added to core 2.0 from the first two rotated cycles
core2set - coreset
```
